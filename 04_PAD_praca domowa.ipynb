Przyjrzyj się danym, wykorzystując poznane metody (describe, info, head) zanalizuj dane.

Następnie:



1. Zamień order_date na format daty. Upewnij się, że w rezultacie data będzie w formacie: rok-miesiąc-dzień;
2. Sprawdź ile i jakie unikalne wartości znajdują się w kolumnie T-shirt category;
3. Wyczyść kolumnę tshirt_category normalizując nazwy produktów. Upewnij się, że dla każdej kategorii funkcjonuje jedna nazwa, np. zamiast white i wh jest tylko jedna wersja white. Skorzystaj z klasy string i metody apply;
4. Rozdziel kolumnę tshirt category na 3 kolumny: tshirt_type, która będzie mówiła o tym czy jest to koszulka, bluza czy coś innego, tshirt_size, która będzie informowała o rozmiarze koszulki, i tshirt_colour, która będzie informowała o kolorze. Skorzystaj z metody apply.
5. Znajdź zamówienia złożone między 31 grudnia 2014 a 2 sierpnia 2016.


```python
import datetime as dt
import pandas as pd
import numpy as np
import csv
df = pd.read_csv('orders.csv',sep = ',')
df.head(15)
# 1.1
df['order_date'] = pd.to_datetime(df['order_date'],format = "%Y.%m.%d")
print('Zadanie 1', '\n', df.head(5), "\n")

# 1.2
print('Zadanie 2', '\n', df['tshirt_category'].value_counts())
print('Tyle unikalnych wartosci = ', len(list(df['tshirt_category'].value_counts())), '\n')

def aplaj(x):
  x = x.split(' ')
  y = len(x)
  z = ''
  for i in range(0,y):
    if x[i] == 'Wh':
      x[i] = 'White'
    if x[i] == 'Bl':
      x[i] = 'Black'
    if x[i] == 'Tshirt':
      x[i] = 'T-Shirt'
    if x[i] == 'Tennis':
      x[i] = 'Tennis-shirt'
    if x[i] == 'Shirt':
      x[i] = ''
    z = z + x[i] + ' '
  return z 

# 1.3
df['tshirt_category'] = df['tshirt_category'].apply(aplaj)
print('Zadanie 3', '\n',df)
# 1.4
def aplaj2(x):
  typ = ['T-Shirt', 'Hoodie', 'Tennis-shirt'] 
  kolor = ['White', 'Black']
  rozmiar = ['M', 'F']

  x = x.split(' ')
  y = len(x)
  if any((match := item) in typ for item in x):
      A = match
  else:
      A = ('-')

  if any((match := item) in rozmiar for item in x):
      B = match
  else:
      B = ('-')

  if any((match := item) in kolor for item in x):
      C = match
  else:
      C = ('-') 
  D = A + ' ' + B  + ' ' + C
  return D

df[['tshirt_Type','tshirt_size','tshirt_colour']] = df['tshirt_category'].apply(aplaj2).str.split(' ',expand = True)
df.drop('tshirt_category', axis = 1, inplace = True)
print('Zadanie 4', '\n',df)

#1.5
df2 = df.loc[df["order_date"].between('2014-12-31', '2016-08-02')]
df2.head(10)
```
